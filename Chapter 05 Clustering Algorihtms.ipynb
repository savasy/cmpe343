````python
kmeans = KMeans(n_clusters=3, random_state=0).fit(X)
````

````python
kmeans.labels_
````

````python
#Q3
'''
Cluster boston dataset into 5 cluster (set random_state=0). What will be the cluster label of the 100th house (please look at labels[100]).
'''

import numpy as np
from sklearn.cluster import KMeans

from sklearn.datasets import load_boston
boston= load_boston()

kmeans = KMeans(n_clusters=5, random_state=0).fit(boston.data)

kmeans.labels_[100]
````

````python
#Q4
'''
Cluster boston dataset into the 5 clusters (set random_state=0). What will be the first value of the first cluster center vector (please see kmeans.cluster_centers_).
'''
import numpy as np
from sklearn.cluster import KMeans

from sklearn.datasets import load_boston
boston= load_boston()

kmeans = KMeans(n_clusters=5, random_state=0).fit(boston.data)
kmeans.cluster_centers_
````

````python
kmeans.cluster_centers_[0,0]
````

````python
#Q5
'''
Reduce boston dataset into 3 components by PCA. And apply kmeans (random_state=0) for 4 clusters. what will be the cluster label of the 10th house (see labels[10]).
'''
from sklearn.cluster import KMeans

from sklearn.datasets import load_boston
boston= load_boston()

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
boston3D=pca.fit_transform(boston.data)

kmeans = KMeans(n_clusters=4, random_state=0).fit(boston3D)
kmeans.labels_[10]
````

Week06 - Test 05

Reduce iris data to 3 dimensions with PCA. And then apply kmeans on reduced iris data for 10 clusters (keep random_state =0). What will be the label of the last flower  

````python
from sklearn.cluster import KMeans

from sklearn.datasets import load_iris
iris= load_iris()

from sklearn.decomposition import PCA
pca = PCA(n_components=3)
iris3D=pca.fit_transform(iris.data)

kmeans = KMeans(n_clusters=10, random_state=0).fit(iris3D)
kmeans.labels_
````

Cluster iris data for 10 clusters (keep random state =0). How many flower will be belongs to label 4 

````python
from sklearn.cluster import KMeans

from sklearn.datasets import load_iris
iris= load_iris()

kmeans = KMeans(n_clusters=10, random_state=0).fit(iris.data)
kmeans.labels_
````

````python
sum(kmeans.labels_==4)
````
