# Q2

Train a missing value imputer (see chapter 03), from boston data set. Use -1 for missing_values and apply mean strategy. Try to predic the value of -1 for the following house.
```
# This is formatted as code
a_home=[-1, 18.0, 2.31, 0.0, 0.538, 6.575, 65.2, 4.09, 1.0, 296.0, 15.3, 396.9, 4.98]

```



````python
import numpy as np
from sklearn.impute import SimpleImputer

imp = SimpleImputer(missing_values=-1, strategy='mean')
imp.fit(boston.data)
````

# Q4
## Q4.A

Apply kmeans to the boston dataset to get 4 clusters (keep random_state as 0). What is the label of the last house.

````python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0).fit(boston.data)
kmeans.labels_
````

````python
df=pd.DataFrame(boston.data, columns=boston.feature_names)
````

````python
df["label"]= kmeans.labels_
````
