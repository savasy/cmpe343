# CH 06 Linear Regression

````python
import numpy as np
from sklearn.linear_model import LinearRegression


X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
X
````

````python
# y = 1 * x_0 + 2 * x_1 + 3
y = np.dot(X, np.array([5, 10])) - 3
y
````

````python
reg = LinearRegression().fit(X, y)
reg
````

````python
reg.coef_
````

````python
reg.intercept_
````

````python
reg.score(X, y)
````

````python
import pandas as pd
df=pd.DataFrame(X, columns=["x0","x1"])
df["y"]=y
df
````
