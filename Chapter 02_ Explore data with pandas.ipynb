# TEST

1) How many rows satisfy the condition "color_intensiy>7 AND hue < 0.7" for the wine dataset

2) How many rows satisfy the condition "ash>2.5 AND magnesium > 80"

3) What is the correlation between alcohol and proline

4) When Sorting mydata by **hue** in ascending order, what would be the value of *magnesium* at the first row ?

5) Change mapping function above so that **"Very Low"** will be defined as if x< 55 (instead of x<40). 
So now, how many category of "very low" we will have.
     




# Solution

```python
# 1) How many rows satisfy the condition "color_intensiy>7 AND hue < 0.7"
mydata[  (mydata.color_intensity>7)   &  (mydata.hue <0.7) ].shape
```

```python
#2) How many rows satisfy the condition "ash>2.5 AND magnesium > 80"
mydata[ (mydata.ash>2.5) & (mydata.magnesium>80)].shape
```

```python
#3) What is the correlation between alcohol and proline

mydata[["alcohol","proline"]].corr()
```

```python
#4) When Sorting mydata by hue in ascending order, what would be the value of magnesium at the first row ?

mydata.sort_values("hue",  ascending=True)['magnesium']
```

```python
# 5) Change mapping function above so that "Very Low" will be defined as if x< 55 (instead of x<40). So now, how many category of "very low" we will have.

def mapping(x):
    if x< 55:
      return "Very Low"
    if x<85:
      return "Low"
    if x<120:
      return "Normal"
    if x <140:
      return "High"
    return "Very High"

mydata["magnesiumCAT"]= mydata.magnesium.map(lambda x: mapping(x))
```
