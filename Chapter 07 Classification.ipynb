Linear Kernel versus Non*linear Kernel

https://github.com/rasbt/python-machine-learning-book-3rd-edition/blob/master/ch03/ch03.ipynb



````python
linear_svc = svm.SVC(kernel='linear')
linear_svc.kernel
````

````python
linear_svc.fit(X,y)
linear_svc.score(X,y)
````

Non-linear kernel

````python
rbf_svc = svm.SVC(kernel='rbf')
rbf_svc.kernel
````

````python
rbf_svc.fit(X,y)
rbf_svc.score(X,y)
````

````python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_circles

# make a synthetic dataset
X, y = make_circles(4000,factor=0.5, random_state=0, noise=0.15)
````

````python
y[:40]
````

````python
plt.scatter(X[:,0], X[:,1], c=y)
````

````python
linear_svc = svm.SVC(kernel='linear')
linear_svc.fit(X,y)
linear_svc.score(X,y)
````

````python
linear_svc = svm.SVC(kernel='rbf')
linear_svc.kernel
linear_svc.fit(X,y)
linear_svc.score(X,y)
````
