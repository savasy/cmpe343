````python
import numpy as np
from sklearn.decomposition import PCA
# Principal Component Analysis

X = np.array([[-1, -1,3], [-2, -1, 3],  [-3, -2, -4], [-2, -1, 13],  [-3, -2, 1], [2, 1, 13],  [-3, 2, 1]])
X
````

````python
pca = PCA(n_components=2)
pca.fit(X)
pca
````

````python
print(pca.explained_variance_ratio_)
````

````python
pca.transform(X)
````

````python
pca.transform([[1,1,1]])
````
